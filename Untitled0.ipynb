```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## Ingreso total por cada tienda
En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_total_tienda1 = sum(tienda['Precio'] + tienda['Costo de envío'])
ingreso_total_tienda2 = sum(tienda2['Precio'] + tienda2['Costo de envío'])
ingreso_total_tienda3 = sum(tienda3['Precio'] + tienda3['Costo de envío'])
ingreso_total_tienda4 = sum(tienda4['Precio'] + tienda4['Costo de envío'])

print(f"Ingreso total Tienda 1: {ingreso_total_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_total_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_total_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_total_tienda4}")
```

## Ventas por categoría
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()

print("Ventas por categoría Tienda 1:")
display(ventas_por_categoria_tienda1)

print("\nVentas por categoría Tienda 2:")
display(ventas_por_categoria_tienda2)

print("\nVentas por categoría Tienda 3:")
display(ventas_por_categoria_tienda3)

print("\nVentas por categoría Tienda 4:")
display(ventas_por_categoria_tienda4)
```

## Calificaciones promedio de los clientes por tienda

Aquí se calcula la calificación promedio de los clientes para cada tienda, utilizando la columna 'Calificación' para determinar la satisfacción del cliente.

```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

## Productos más y menos vendidos por tienda

En esta sección, se identificarán los 5 productos más vendidos y los 5 menos vendidos para cada tienda, basándose en la frecuencia de aparición de cada producto en los registros de ventas.

```python
# Calcular la frecuencia de ventas de cada producto en cada tienda
ventas_producto_tienda1 = tienda['Producto'].value_counts()
ventas_producto_tienda2 = tienda2['Producto'].value_counts()
ventas_producto_tienda3 = tienda3['Producto'].value_counts()
ventas_producto_tienda4 = tienda4['Producto'].value_counts()

# Mostrar los 5 productos más vendidos y los 5 menos vendidos para cada tienda
print("\n--- Tienda 1 ---")
print("5 Productos más vendidos:")
display(ventas_producto_tienda1.head(5))
print("\n5 Productos menos vendidos:")
display(ventas_producto_tienda1.tail(5))

print("\n--- Tienda 2 ---")
print("5 Productos más vendidos:")
display(ventas_producto_tienda2.head(5))
print("\n5 Productos menos vendidos:")
display(ventas_producto_tienda2.tail(5))

print("\n--- Tienda 3 ---")
print("5 Productos más vendidos:")
display(ventas_producto_tienda3.head(5))
print("\n5 Productos menos vendidos:")
display(ventas_producto_tienda3.tail(5))

print("\n--- Tienda 4 ---")
print("5 Productos más vendidos:")
display(ventas_producto_tienda4.head(5))
print("\n5 Productos menos vendidos:")
display(ventas_producto_tienda4.tail(5))
```

## Visualización de Productos más y menos vendidos

A continuación, se presentan gráficos de barras para visualizar los 5 productos más y menos vendidos en cada tienda, facilitando la identificación de tendencias.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Función auxiliar para graficar productos
def plot_top_bottom_products(product_counts, title_suffix, store_name):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    fig.suptitle(f'Productos en {store_name} - {title_suffix}', fontsize=16)

    # Productos más vendidos
    sns.barplot(x=product_counts.head(5).index, y=product_counts.head(5).values, ax=axes[0], palette='viridis')
    axes[0].set_title('5 Productos Más Vendidos')
    axes[0].set_xlabel('Producto')
    axes[0].set_ylabel('Número de Ventas')
    axes[0].tick_params(axis='x', rotation=45)

    # Productos menos vendidos
    sns.barplot(x=product_counts.tail(5).index, y=product_counts.tail(5).values, ax=axes[1], palette='magma')
    axes[1].set_title('5 Productos Menos Vendidos')
    axes[1].set_xlabel('Producto')
    axes[1].set_ylabel('Número de Ventas')
    axes[1].tick_params(axis='x', rotation=45)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# Graficar para cada tienda
plot_top_bottom_products(ventas_producto_tienda1, '', 'Tienda 1')
plot_top_bottom_products(ventas_producto_tienda2, '', 'Tienda 2')
plot_top_bottom_products(ventas_producto_tienda3, '', 'Tienda 3')
plot_top_bottom_products(ventas_producto_tienda4, '', 'Tienda 4')
```

## Costo de envío promedio por tienda

En esta sección, se calculará el costo de envío promedio para cada tienda, utilizando la columna 'Costo de envío' para entender cuánto se gasta, en promedio, en el envío por cada tienda.

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio Tienda 1: {costo_envio_promedio_tienda1:.2f}")
print(f"Costo de envío promedio Tienda 2: {costo_envio_promedio_tienda2:.2f}")
print(f"Costo de envío promedio Tienda 3: {costo_envio_promedio_tienda3:.2f}")
print(f"Costo de envío promedio Tienda 4: {costo_envio_promedio_tienda4:.2f}")
```

## Visualizaciones Clave de los Análisis

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Gráfico de Ingreso Total por Tienda
ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Ingreso Total', data=ingresos_totales, palette='viridis')
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')
plt.ticklabel_format(style='plain', axis='y') # Evitar notación científica en el eje Y
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 2. Gráfico de Calificación Promedio de Clientes por Tienda
calificaciones_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Calificación Promedio', data=calificaciones_promedio, palette='plasma')
plt.title('Calificación Promedio de Clientes por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5) # Establecer el rango del eje Y de 0 a 5 para calificaciones
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3. Gráfico de Ventas por Categoría de Producto por Tienda

# Combinar los datos de ventas por categoría de todas las tiendas en un solo DataFrame
ventas_categorias_df = pd.DataFrame({
    'Categoría del Producto': ventas_por_categoria_tienda1.index,
    'Tienda 1': ventas_por_categoria_tienda1.values,
    'Tienda 2': ventas_por_categoria_tienda2.values,
    'Tienda 3': ventas_por_categoria_tienda3.values,
    'Tienda 4': ventas_por_categoria_tienda4.values
})

# Reestructurar el DataFrame para Seaborn (formato 'long')
ventas_categorias_melted = ventas_categorias_df.melt(id_vars='Categoría del Producto', var_name='Tienda', value_name='Número de Ventas')

plt.figure(figsize=(14, 8))
sns.barplot(x='Categoría del Producto', y='Número de Ventas', hue='Tienda', data=ventas_categorias_melted, palette='tab10')
plt.title('Número de Ventas por Categoría de Producto y Tienda')
plt.xlabel('Categoría del Producto')
plt.ylabel('Número de Ventas')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
